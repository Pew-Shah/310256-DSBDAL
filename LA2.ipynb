```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
'Firstname': ["Ananya", "Arjun", "Aarav", "Aditi", "Aryan", "Akshay", "Bhavna", "Dev", "Diya", "Gautam", "Ishika", "Kunal"],
    'Lastname': ['Roy', 'Dixit', 'Danial', 'Kapoor', 'Jha', 'Thakur', 'Kapoor', 'Khanna', 'Pardesi', 'Pardesi', 'Desai', 'Patel'],
    'Gender': ['F', 'M', 'M', 'F', 'M', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'SPOS': [85.0, 90.0, np.nan, 95.0, np.nan, 87.0, 81.0, 40.0, 10.0, 9.0, 200.0, 300.0],
    'DSBDA': [88, 90, 'na', 86, 84, np.nan, 80, 88, 96, np.nan, 87, np.nan],
    'WT': [90.0, 94.0, 100.0, 500.0, 98.0, 97.0, 96.0, 95.0, 94.0, 93.0, np.nan, 88.0],
    'DA': [92, 95, 80, 82, 84, 86, 89, 87, 88, 96, 'N.A', 300]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
missing_value=['NaN','na','N.A']
df.replace(missing_value,np.nan,inplace=True)
df
```

```python
df2=df.dropna()
df2
```

```python
df['SPOS'].fillna(method='pad',inplace=True)
df
```

```python
df.WT.fillna(method='bfill',inplace=True)
df
```

```python
sns.boxplot(y=df['SPOS'])
plt.show()
```

```python
sns.boxplot(data=df,x='WT')
```

```python
df.SPOS=np.where(df.SPOS>100,df.SPOS.median(),df.WT)
df
```

```python
sns.boxplot(data=df,x='SPOS')
```

```python
df['DSBDA'].fillna(method='pad',inplace=True)
df
```

```python
df['DSBDA_log']=np.log(df.DSBDA)
```

```python
#log transformation
df.DSBDA
```

```python
#log transformation
df['DSBDA_log']
```

```python
sns.histplot(df.DSBDA_log, kde=True)
plt.title('Decreasing skewness and converting the distribution into a normal distribution')
plt.show()
#This transformation can reduce right skewness and make the distribution more symmetrical.
```

```python
#sqrt transformation
df['DSBDA_sq']=np.sqrt(df.DSBDA)
print(df.DSBDA,df.DSBDA_sq)
```

```python
sns.histplot(df.DSBDA_sq, kde=True)
plt.title('Converting a non-linear relation into a linear one')
plt.show()
#This can help linearize the relationship between 'SPOS' and other variables if the original relationship is non-linear.
```

```python
#scale transformation
min_val = df['DSBDA'].min()
max_val = df['DSBDA'].max()
df['DSBDA_minmax'] = (df['DSBDA'] - min_val) / (max_val - min_val)
```

```python
df['DSBDA_minmax']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming your DataFrame is named 'df' and your variable is 'DSBDA'
scaler = MinMaxScaler(feature_range=(0, 1))  # Specify range (default 0-1)
df['DSBDA_scaled'] = scaler.fit_transform(df[['DSBDA']])  # Fit and transform for scaling

df['DSBDA_scaled']
```

```python
sns.histplot(df.DSBDA_scaled, kde=True)
plt.title('Changing the scale for better understanding of the variable')
plt.show()
#We can apply min-max scaling to rescale the values of 'SPOS' to a range between 0 and 1. This can make the variable more interpretable by bringing all values into a common scale.
```
