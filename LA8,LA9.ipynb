```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=sns.load_dataset('titanic')
ds.head()
```

```python
ds.isnull().sum()
```

```python
ds.age=ds.age.fillna(np.mean(ds.age))
```

```python
ds.age.isnull().sum()
```

```python
ds.deck=ds.deck.fillna(ds.deck.mode()[0])
```

```python
ds.deck.isnull().sum()
```

```python
ds.embark_town=ds.embark_town.fillna(ds.embark_town.mode()[0])
```

```python
ds.embark_town.isnull().sum()
```

```python
plt.figure(figsize=(3,5))
sns.countplot(x='sex',data=ds)
plt.title('Passenger Count by Sex')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='pclass',hue='sex',data=ds)
plt.title('Passenger Count by Pclass and Sex')
plt.show()
```

```python
plt.figure(figsize=(5,2))
sns.histplot(ds['age'],bins=10,kde=True)
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(5,2))
sns.barplot(x='pclass',y='survived',data=ds)
plt.ylabel('survival rate')
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
sns.countplot(x='pclass',hue='survived',data=ds)
plt.title('Survival Rate by Pclass WITH COUNTPLOT')
```

```python
#LA8 b
sns.histplot(x='fare',bins=20,data=ds)
plt.title('Distribution of Ticket Prices')
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.boxplot(x='pclass',y='age',hue='survived',data=ds)
plt.title('LA9 A')
plt.show()
```

```python
sns.jointplot(x='age',y='fare',data=ds)
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(y='age',x='sex',hue='survived',data=ds)
plt.show()
```
